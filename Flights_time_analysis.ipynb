```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
# dataset load
sns.load_dataset("flights")
```

```python
df = sns.load_dataset("flights")
df.shape
```

```python
df.info()
```

```python
df.head()
```

# 다양한 시각화 해보기

```python
# barplot
plt.figure(figsize=(15,4))
sns.barplot(data=df, x='year', y='passengers')
```

```python
# lineplot
plt.figure(figsize=(15,4))
sns.lineplot(data=df, x='year', y='passengers')
```

```python
# point plot

plt.figure(figsize=(15,4))
sns.pointplot(data=df, x='year', y='passengers', ci='sd') # ci sd는 신뢰구간을 표준편차로 씀을 의미함
```

```python
# barplot
plt.figure(figsize=(15,4))
sns.barplot(data=df, x='year', y='passengers', hue='month')
```

```python
# lineplot + lenend(범례) 멀찍이 띄우기 bbox_to_anchor=(x,y) x로 가로조절

plt.figure(figsize=(15,4))
sns.lineplot(data=df, x='year', y='passengers', hue='month')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# hue 값 변경

plt.figure(figsize=(15,4))
sns.lineplot(data=df, x='month', y='passengers', hue='year')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# box plot


plt.figure(figsize=(15,4))
sns.boxplot(data=df, x='year', y='passengers')
#plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.) # hue가 있을 때 legend사용이 들어오구나!
```

```python
# boxen plot
plt.figure(figsize=(15,4))
sns.boxenplot(data=df, x='year', y='passengers')
```

```python
# violin plot
plt.figure(figsize=(15,4))
sns.violinplot(data=df, x='year', y='passengers') # 흰색 point는 median을 뜻함
```

```python
# swarmplot 퍼뜨려 주구나 !
plt.figure(figsize=(15,4))
sns.swarmplot(data=df, x='year', y='passengers')
```

```python
# box (monthly)
plt.figure(figsize=(15,4))
sns.boxplot(data=df, x='month', y='passengers')
```

```python
# reg plot

plt.figure(figsize=(15,4))
sns.regplot(data=df, x='year', y='passengers') # x는 integer야 함!
```

```python
# swarm plot

plt.figure(figsize=(15,4))
sns.swarmplot(data=df, x='year', y='passengers') # x는 integer야 함!
```
